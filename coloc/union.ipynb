```python
import pandas as pd
path='/gpfs/commons/groups/knowles_lab/data/ADSP_reguloML/PRS/36k_hg38/snp_pip_thres/remove_index0/'
susie = pd.read_csv(path + 'bellenguez_susie.txt', sep = '\t')
baseline = pd.read_csv(path + 'bellenguez_baseline.txt', sep = '\t')
omics  = pd.read_csv(path + 'bellenguez_omics.txt', sep = '\t')
dl= pd.read_csv(path + 'bellenguez_omics_dl.txt', sep = '\t')
mutual=pd.read_csv(path+'mutual_snp.tsv ', sep = '\t')
```

```python
def get_region(df,anno):
    df['Region'] = df['CREDIBLE_SET'].str.extract(r'chr\d+:(\d+)-(\d+)').astype(int).mean(axis=1).astype(int)    
    df['source'] = anno    
    return(df.loc[:,['CHR', 'BP','SNP','CREDIBLE_SET','Region', 'source']])
        
    
```

```python
union = pd.concat([get_region(susie,'susie'),get_region(baseline,'baseline'), get_region(omics,'omics'), get_region(dl, 'omics_dl')])
```

```python
union.to_csv(path+'union_snp_of_interest_info.tsv',sep = '\t', index = False)
union
```

```python
snp_of_interest = union.loc[:,['CHR','source','Region']].drop_duplicates()
snp_of_interest.to_csv(path+'union_snp_of_interest.tsv',sep = '\t', index = False)
```
