```python
import math
import pandas as pd
import numpy as np
```

```python
hg38 = pd.read_csv('/gpfs/commons/groups/knowles_lab/data/ADSP_reguloML/summary_stats/alzheimers/fixed_alzheimers/processed/bellenguez_hg38_from_parquet.tsv', sep = '\t')
```

```python
## check what column we need
## cojo needs SNP A1 A2 freq b se p N 
hg38
```

```python
hg38['se'] = 1/np.sqrt(2*hg38.N*hg38.MAF*(1-hg38.MAF))
```

```python
hg38['b'] = hg38.Z * hg38.se
```

```python
from scipy.stats import norm
hg38['p'] = 2 * (1 - norm.cdf(np.abs(hg38.Z)))
```

```python
hg38 = hg38.rename(columns = {"MAF":'freq'})
```

```python
##SNP A1 A2 freq b se p N 
save = hg38.loc[:,["SNP", "A1", "A2", "freq", "b" ,"se" ,"p" ,"N"]]
save
```

```python
save.to_csv('/gpfs/commons/groups/knowles_lab/data/ADSP_reguloML/summary_stats/alzheimers/fixed_alzheimers/processed/sbayesrc/bellenguez_hg38_from_parquet.cojo', sep = '\t', index = False)
            
```
