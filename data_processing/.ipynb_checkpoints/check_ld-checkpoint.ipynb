```python
def plot_ld(start, end, ld):
   
    file = '/gpfs/commons/groups/knowles_lab/data/ldsc/polyfun/ukb_ld/'+ ld
    LD = np.load(file+".npz")
    LD_info = pd.read_csv(file+'.gz', compression='gzip', sep = '\t')
    shape = LD["row"][-1]+2
    row = LD["row"]
    column = LD["col"]
    value = LD["data"]
    
    coo = sparse.coo_matrix((value, (row, column)), shape=(shape,shape))
    coo_dense = coo.todense()
    if(str(start) in ld and str(end) in ld):
        index = range(0, shape, 1)
    else:
        index = LD_info[(LD_info.position >= start) & (LD_info.position <= end)].index.values.tolist()
    
    plt.figure(figsize=(20, 20))

    mask = np.zeros_like(coo_dense[np.ix_(index,index)])
    mask[np.triu_indices_from(mask)] = True



    with sns.axes_style("white"):
        sns.set(font_scale=1.1)
        ax = sns.heatmap(coo_dense[np.ix_(index,index)], mask=mask, square=True,  cmap="gist_rainbow_r", vmin=-1, vmax=1)
        ax.set(title="bellenguez_"+ str(start)+"_"+ str(end) )
        #plt.savefig("/gpfs/commons/home/tlin/pic/LD_heatmap_notransparant")
        #plt.savefig("/gpfs/commons/home/tlin/pic/LD_heatmap", transparent=True)
        plt.show()
    
```

```python
plot_ld(6000001,7500001,"chr1_6000001_9000001")
```

```python
plot_ld(20500001,22000001,"chr1_20000001_23000001")
```
